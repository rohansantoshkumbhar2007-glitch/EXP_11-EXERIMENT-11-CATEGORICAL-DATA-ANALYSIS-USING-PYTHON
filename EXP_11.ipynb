```
import pandas as pd
data = {
    'Order_ID':['01','02','03','04','05','06','07','08','09','010'],
    'Category':['Eletonics','Clothing','Eletonics','Grocery','Clothing','Eleronics','Grocery','Clothing','Electronics','Grocery'],
    'Payemnt_Method':['UPI','Card','UPI','Cash','Card','UPI','Cash','Card','UPI','Cash'],
    'Delivery_Type':['Express','Standard','Express','Standard','Express','Standard','Express','Standard','Express','Standard'],
    'Customer_Type:':['New','Returning','Returning','New','Returning','New','Returning','New','Returning','New']


}
df=pd.DataFrame(data)
print(df)
```

```
#1. Frequency Count
#Count how many orders in each category

print(df['Category'].value_counts())
```

```
#3. Number of Unique Values
print(df['Category'].nunique())
```

```
#4. Frequecy of Payment Methods
print(df['Payemnt_Method'].value_counts())
```

```
#5. cross Tabulation
#category vs Payment Method

print(pd.crosstab(df['Category'],df['Payemnt_Method']))
```

```
#6. Cross Tabulation (Delivery vs Customer Type)
print(pd.crosstab(df['Delivery_Type'],df['Customer_Type:']))
```

```
#7. Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```
#8.Filtering Data
#Example:Orders from Eletronics category
electronics_orders=df[df['Category']=='Eletonics']
print(electronics_orders)
```

```
#8.Filtering Data
#Example:Orders from Eletronics category
electronics_orders=df[df['Category']=='Clothing']
print(electronics_orders)
```

```
#9.Grouping
#Orders by category and payment method
print(df.groupby(['Category','Payemnt_Method']).value_counts())
```
